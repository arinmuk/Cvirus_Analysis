```python
def cleanup (cvdf1):
    cvdf1['Deaths']=cvdf1.Deaths.fillna(0)
    cvdf1['Confirmed']=cvdf1.Confirmed.fillna(0)
    #cvdf['Deaths']=cvdf.Deaths.fillna(0)
    cvdf1['Recovered']=cvdf1.Recovered.fillna(0)
    cvdf1['Active']=cvdf1.Active.fillna(0)
    cvdf1['Case_Fatality_Ratio']=cvdf1.Case_Fatality_Ratio.fillna((cvdf1['Deaths']/cvdf1['Confirmed'])*100)
    cvdf1['Combined_Key']=cvdf1.Combined_Key.fillna(cvdf1['Province_State'] + ',' + cvdf1['Country_Region'])
    sqldf1 = cvdf1.copy()
    #sqldf = cvdf.head(20000)
    #sqldf.drop("_id",axis=1)
    lstckey = sqldf1['Combined_Key'].unique()
    lstckey
    sqljoindf1=pd.DataFrame({})
    sqldf1=sqldf1.sort_values(by=['Combined_Key',"Last_Update"])
    for country in lstckey:
        print(country)
        workframe1=pd.DataFrame({})
        workframe1 = sqldf1[(sqldf1.Combined_Key==country)]
        workframe1["daily_conf"]=workframe1['Confirmed'].diff()
        workframe1["daily_deaths"]=workframe1['Deaths'].diff()
        workframe1["daily_recov"]=workframe1['Recovered'].diff()

        sqljoindf1= pd.concat([sqljoindf1,workframe1])
    return cvdf1 , sqljoindf1
cvdf2,sqljoindf2 = cleanup (workingdf)
```
